- 252135 values in the DAYS_EMPLOYED column were negative and 55374 values were positivein DAYS_EMPLOYED column in application_train. This looks like an error in data collecton. So I decided to take absolute of values in thsi column and then divide by 365 to change it into YEARS_EMPLOYED.

```python
df1= application_train                                               #Create new dataset from application_train

df1['YEARS_BIRTH'] = (application_train['DAYS_BIRTH'].abs())/365           #Change DAYS_BIRTH column to YEARS_BIRTH
df1 = df1.drop(['DAYS_BIRTH'], axis=1)

df1['YEARS_EMPLOYED'] = (application_train['DAYS_EMPLOYED'].abs())/365    #Change DAYS_EMPLOYED column to YEARS_EMPLOYED
df1 = df1.drop(['DAYS_EMPLOYED'], axis=1)

df1.head()
```
